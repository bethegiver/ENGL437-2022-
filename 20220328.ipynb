```python
def test(dogcat):
  URL = ('http://apis.data.go.kr/1543061/abandonmentPublicSrvc/'
        'abandonmentPublic?'
        '_type=json'
        '&bgnde=20210101'
        '&endde='+dogcat+
        '&pageNo=1'
        '&numOfRows=100'
        '&serviceKey=bQfA1NhnYvKzvrDM85mtrN9DP8ldfxPOtaJEwRCMAM1kSA8YXXOIdPCj9%2B7GmmbYOlpfH3aQiAwhizuJ4hUE6w%3D%3D')
  response = requests.get(URL)
  r = response.json()
  items = r['response']['body']['items']['item']

  processState = [item['processState'] for item in items]
  age = [item['age'] for item in items]
  happenPlace = [item['happenPlace'] for item in items]
  kindCd = [item['kindCd'] for item in items]
  specialMark = [item['specialMark'] for item in items]
  weight = [item['weight'] for item in items]
  popfile = [item['popfile'] for item in items]

  processState = pd.Series(processState)
  age = pd.Series(age)
  happenPlace = pd.Series(happenPlace)   
  kindCd = pd.Series(kindCd)
  specialMark = pd.Series(specialMark)
  weight = pd.Series(weight)
  popfile = pd.Series(popfile)

  df = pd.DataFrame({ 'processState' : processState, 'age': age, 'happenPlace': happenPlace, 'kindCd': kindCd, 'specialMark': specialMark, 'weight': weight, 'popfile' : popfile})
  return df
```

```python
test('20210104')
```

```python
iface = gr.Interface(test,
                     'text',
                     'dataframe',
                     examples = ['input a date (ex. 20210131). It will give you the dataframe of lost dogs and cats until that date from 2021.01.01.']
                     )
iface.launch(debug=True)
```
